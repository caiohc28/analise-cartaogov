```python
import pandas as pd

df = pd.read_csv('dataframe.csv', sep=';', encoding='latin1')

display(df)
```

```python
df = df.dropna()

display(df.info())
```

```python
df.drop(columns=['CÓDIGO ÓRGÃO SUPERIOR', 'CÓDIGO UNIDADE GESTORA', 'CÓDIGO ÓRGÃO'])
```

```python
from sklearn.preprocessing import LabelEncoder

valor_trans = LabelEncoder()
df["VALOR TRANSAÇÃO"] = valor_trans.fit_transform(df["VALOR TRANSAÇÃO"])

display(df.info())
```

```python
display(df['VALOR TRANSAÇÃO'].value_counts())
```

```python
df.groupby('NOME ÓRGÃO SUPERIOR')['VALOR TRANSAÇÃO'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('NOME ÓRGÃO SUPERIOR')['VALOR TRANSAÇÃO'].sum().sort_values(ascending=False).head(10).plot(kind='barh', figsize=(10, 8), title='Top 10 Órgãos Superiores por Valor de Transação (em milhões)')
```

```python
df.groupby('TRANSAÇÃO')['VALOR TRANSAÇÃO'].sum().sort_values(ascending=False)
```

```python
import plotly.express as px

for orgao in df['NOME ÓRGÃO SUPERIOR'].unique():

    gastos = df[df['NOME ÓRGÃO SUPERIOR'] == orgao].groupby('TRANSAÇÃO')['VALOR TRANSAÇÃO'].sum().sort_values(ascending=False)

    grafico = px.bar(
        x=gastos.index,
        y=gastos.values,
        title=orgao,
        text_auto= 'True'
    )

    grafico.show()
```
